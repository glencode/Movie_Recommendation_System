```python
import numpy as np
import pandas as pd
import difflib# For finding the closest matche
from sklearn.feature_extraction.text import TfidfVectorizer# For creating a TF-IDF matrix
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_dataset = pd.read_csv('data/movies.csv')
```

```python
movies_dataset.head()
```

```python
movies_dataset.shape
```

```python
movies_dataset.columns
```

```python
#selecting the features
features = ['keywords','cast','genres','director', 'vote_average', 'overview']

print(features)
```

```python
#replace the NaN values with empty string
for feature in features:
    movies_dataset[feature] = movies_dataset[feature].fillna('')
```

```python
#combining all the selected features
combined_features = movies_dataset['genres']+''+movies_dataset['keywords']+''+movies_dataset['cast']+''+movies_dataset['director']+''+movies_dataset['overview']+''+movies_dataset['vote_average'].astype(str)
```

```python
print(combined_features)
```

```python
#converting the combined features into numeric vectors
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
#calculating the cosine similarity********************************
cosine_sim = cosine_similarity(feature_vectors)
```

```python
print(cosine_sim)
```

```python
print(cosine_sim.shape)
```

```python
#getting the name of the movie from the user
movie_name = input("Enter your favourite movie name: ")
```

```python
#creating a list with all the movie names in the dataset
movie_list = movies_dataset['title'].tolist()
print(movie_list)
```

```python
#finding the closest match to the movie name entered by the user
match = difflib.get_close_matches(movie_name, movie_list, n=1, cutoff=0.6)
print(match)
```

```python
closest_match = match[0]#getting the first match
print(closest_match)
```

```python
#finding the index of the movie in the dataset
movie_index = movies_dataset[movies_dataset.title == closest_match].index[0]
print(movie_index)
```

```python
#preparing the list of similar movies
similar_movies = list(enumerate(cosine_sim[movie_index]))
print(similar_movies)
```

```python
#length of the list
len(similar_movies)
```

```python
#sort by the truly similar movies
sorted_similar_movies = sorted(similar_movies, key=lambda x:x[1], reverse=True)
print(sorted_similar_movies)	
```

```python
#selecting the top 20 similar movies 
# top_20_similar_movies = sorted_similar_movies[1:21]
print('Movies suggested for you are: \n')

i = 1#counter

for movie in sorted_similar_movies:#looping through the sorted list
    print(str(i) + '. ' + movies_dataset['title'][movie[0]])#printing the movie name
    i = i+1#incrementing the counter
    if i>20:#if the counter reaches 20, break the loop
        break#break the loop
```
